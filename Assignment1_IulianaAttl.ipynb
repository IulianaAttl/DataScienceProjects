**What is pandas?**

>Pandas is a library in python whose features can be described as similar to Microsoft Excel. The data structure used for pandas is DataFrame which is made up of a collection of series. When data is loaded into the library it is stored in a csv file and manipulated by the DataFrame which supports data types such as lists and arrays. DataFrame is tabular, meaning that data is arranged in rows (observations) and columns (variables).

**Why use pandas?**

>Pandas gives users the possibility of using python language to manipulate larger data without limitations such as size. It takes unstructured raw data and transforms it into readable and structured data. 

```
#create animals data
al = ['Elephant', 'Giraffe', 'Grizzly_Bear', 'Ostrich', 'Kangaroo', 'King_Cobra', 'Beaver', 'Crocodile', 'Amur_Leopard', 'Hyena', 'Tiger', 'Black_Rhino', 'Sea_Lion', 'Green_Turtle']
dt = ['Herbivore', 'Herbivore', 'Omnivore', 'Omnivore', 'Herbivore', 'Carnivore', 'Herbivore', 'Crnivore', 'Carnivore', 'Carnivore', 'Carnivore', 'Herbivore', 'Carnivore', 'Herbivore']
ctry = ['China', 'Africa', 'Alaska', 'Africa', 'Australia', 'India', 'America', 'Australia', 'Russia', 'Africa', 'China', 'Africa', 'America', 'Alaska']
an = [30000, 68000, 55000, 150000, 50000000, 8000, 1221130, 100000, 84, 14000, 3900, 5500, 18000, 90000]
ls = [48, 26, 25, 60, 6, 20, 15, 70, 15, 12, 15, 50, 30, 70]
er = [True, False, False, False, False, True, False, False, True, False, True, True, True, True]

#create a dictionary
animal_dict = {'animal':al, 'diet':dt, 'country':ctry, 'approx_num':an, 'lifespan':ls, 'endangered':er}

#import pandas
import pandas as pd

#create the DataFrame
animals = pd.DataFrame(animal_dict)

#display the DataFrame
display(animals)
```

**Method 1**

groupby() : this method allows you to split the data into different groups, apply a function if you want and then combine the groups back into one

Reference:

https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html

```
#see information about the endangered animals grouped by each country

#get column endangered
edgr = animals['endangered']

#get just those that are true
endagered_animals = animals[edgr]

#remove the country column as country is already displayed as each group's heading
endagered_animals = endagered_animals.drop('country', axis=1)

#can also remove the endangered column as the result gives the endangered animals
endagered_animals = endagered_animals.drop('endangered', axis=1)

#group the results by country
grouped_animals = endagered_animals.groupby(animals['country'])

#display the groups
for rownum, row in grouped_animals: 
    display(rownum)  #row number
    display(row)  #row data
```

**Method 2**

sort_values() : this method can be used on both series and DataFrame to sort data by a column or row

Reference:

https://pandas.pydata.org/pandas-docs/stable/user_guide/basics.html

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sort_values.html 

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.reset_index.html

```
#sort the DataFrame by the lifespan of animals
sorted_animals = animals.sort_values(by = 'lifespan')
display(sorted_animals)
```

```
#after sorting by lifespan the index or row number is not in ascending order
#want to keep this order of animals by the lifespan

#reset the index but keep everything in place
sorted_animals.reset_index(drop = True, inplace = True)
display(sorted_animals)
```

```
#original DataFrame
display(animals)
```

rename() : this method allows you to change values in the DataFrame such as row labels or column names

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html

https://pandas.pydata.org/pandas-docs/stable/user_guide/basics.html

**Method 4**

nlargest() - returns the number of rows specified with the largest values in the column specified

nsmallest() - return the number of rows specified with the smallest values in the column specified

References:

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.nlargest.html

```
#find the 4 largest numbers of animals left
largest_number = animals.nlargest(4, 'approx_num')
display(largest_number)
```

```
#find the 4 smallest numbers of animals left
smallest_number = animals.nsmallest(4, 'approx_num')
display(smallest_number)
```

**Method 5**

where() : this method checks for a condition and if it is not true it hides the data on the row where it is not true

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.mask.html

```
#if i want to count how many herbivores are in this DataFrame  - if you do this on the original animals DataFrame it will change it
#check for Herbivores in the diet column
herbivores = animals_copy['diet'] == 'Herbivore'

#see where the condition is true
animals_copy.where(herbivores, inplace=True)
display(animals_copy)
```
