### **Overfit & Underfit**

Overfit if taking too many points into account. Underfit is when not enough points are taken into account. For overfit the accuracy is too high and for underfit the accuracy is too low. The best is where the sweetspot is. This is where the training data and the tests data are closest together and move at the same pace. 

From looking at the below example from using the wine dataset we can see that the sweetspot is around 4 or 5 number of neighbours. 

### **Part 5**

```
#import libraries
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import numpy as np
from matplotlib.colors import ListedColormap
from sklearn.neighbors import NearestCentroid 
from matplotlib.ticker import MaxNLocator
from collections import namedtuple
import timeit
```

```
#load the wine dataset
wine = datasets.load_wine()
X = wine.data
y = wine.target
```

### **Nearest Neighbours**

```
#pass the complete data X and y, test size should be 30%, random state - randomly shuffle this by 60 so each time it is repeated its gonna be the same
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=60, stratify=y) 
#see results for 10 neighbours
neighbors = np.arange(1, 10)
# arrays storing train and test accuracies
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

#for each value of k set up a classifier,fit it onto the training data and compute accuracies
for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train_accuracy[i] = knn.score(X_train, y_train) 
    test_accuracy[i] = knn.score(X_test, y_test)

#plot to see number of neighbours
plt.title('k-NN: Varying Number of Neighbours')
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('Number of Neighbours')
plt.ylabel('Accuracy')
plt.show()
```

### **Nearest Centroid**

```
#create colour maps
cmap_light = ListedColormap(['#FF9C9C', '#83d8fc', '#f7f0b0'])
cmap_dark = ListedColormap(['#FA4343', '#1CBBFF', '#f7e64a'])
#subset the data set to just 2 attributes to see the decision plot
X = wine.data[:, :2] 
y = wine.target

#step size
h = .01

for shrinkage in [None, .5, 1.0, 1.5, 2.0]:
    nc = NearestCentroid(shrink_threshold=shrinkage)
    nc.fit(X, y)
    y_pred = nc.predict(X) 
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    #centroid prediction
    Z = nc.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    #plot the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_dark, edgecolor='k', s=20, alpha=0.65)
    plt.scatter(nc.centroids_[:, 0], nc.centroids_[:, 1], c=[0, 1, 2], cmap=cmap_dark, edgecolor='k', s=160, marker='*')
    plt.title("3-Class classification (shrink_threshold=%r)" % shrinkage)
    plt.axis('tight')

plt.show()
# http://statweb.stanford.edu/~tibs/PAM/Rdist/howwork.html
```

For this we can conclude that the bigger the parameter or shrinkage the tighter together they become.

```
# Setup arrays to store train and test accuracies
shrinkage_param_list = [0.0, 0.2, 0.4, 0.6, 0.8]
train_accuracy = np.empty(len(shrinkage_param_list))
test_accuracy = np.empty(len(shrinkage_param_list))

#for every item of k
for i, shrinkage in enumerate(shrinkage_param_list):
    nc = NearestCentroid(shrink_threshold=shrinkage)
    nc.fit(X_train, y_train)
    train_accuracy[i] = nc.score(X_train, y_train)
    test_accuracy[i] = nc.score(X_test, y_test)
```

```
#calculate the time it took to run
number = 2
repeat = 4

#make copies of the data
def vcopies(arr, num=4):
    return np.vstack( (arr for i in range(num)) ) 

def hcopies(arr, num=4):
    return np.hstack( (arr for i in range(num)) )

X_train2 = vcopies(X_train) 
y_train2 = hcopies(y_train)
X_test2 = vcopies(X_test)
y_test2 = hcopies(y_test)

#nearest neighbour training time
knn_train_time = timeit.repeat('knn = KNeighborsClassifier(n_neighbors=1, algorithm="kd_tree"); knn.fit(X_train2, y_train2)',globals=globals(), number=number, repeat=repeat)
#nearest centroid training time
nc_train_time = timeit.repeat('nc = NearestCentroid(); nc.fit(X_train2, y_train2)',globals=globals(), number=number, repeat=repeat)
#nearest neighbour prediction time
knn_predict_time = timeit.repeat('knn.predict(X_test2)', globals=globals(), number=number, repeat=repeat)
#nearest centroid prediction time
nc_predict_time = timeit.repeat('nc.predict(X_test2)', globals=globals(), number=number, repeat=repeat)

#get the average time
def get_avg_time(time_take):
    return np.array(time_take).mean()/number

#calculate the average time for each 
knn_train_avg_time = get_avg_time(knn_train_time)
knn_predict_avg_time = get_avg_time(knn_predict_time)
nc_train_avg_time = get_avg_time(nc_train_time)
nc_predict_avg_time = get_avg_time(nc_predict_time)
```

```
#groups for the graph
n_groups = 2

knn_perf = (knn_train_avg_time, knn_predict_avg_time)
nn_perf = (nc_train_avg_time, nc_predict_avg_time)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.40

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = ax.bar(index, knn_perf, bar_width, alpha=opacity, color='g', label='kNN')
rects2 = ax.bar(index + bar_width, nn_perf, bar_width, alpha=opacity, color='r', label='NC')

ax.set_ylabel('Time taken')
ax.set_title('Performance')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(('Train', 'Test'))
ax.legend()

plt.show()
```

As we can see from the graph nearest neighbour takes longer on the training and testing than nearest centroid. This means that nearest centroid is best for classification.
